# mean v_a vs noise 

```python
# NOTICE: input is an array [mean_v_a,rho, eta]
def plot_va_eta(eta_matrix, shape, color, labels):
    # Plot mean order parameter as a function of density when noise is fixed
    plt.title("Order Parameter vs Noise")
    plt.xlabel("η")
    plt.ylabel("$v_a$")
    plt.plot(eta_matrix[:, 2], eta_matrix[:, 0], shape, markerfacecolor='none', markeredgecolor=color, label=labels)
    plt.legend()


plot_va_eta(bird_sim1.eta_matrix, 's','#355070','N = 40') 
plot_va_eta(bird_sim2.eta_matrix, 'x','#6d597a','N = 100')
plot_va_eta(bird_sim3.eta_matrix,'+' ,'#b56576', 'N = 400')
plot_va_eta(bird_sim4.eta_matrix,'^' ,'#214e34', 'N = 1000')
#plot_va_eta(bird_sim4.eta_matrix,'d' ,'#eaac8b', 'N = 4000')
plt.show()
```

# v_a vs time for N = 40 with various noise 

```python
import matplotlib.pyplot as plt
import dill
import numpy as np

name = input('Desired file name identification?\t')
with open( name + '_fixrhoeta_swarm.csv',"rb") as pickle_in:  # "rb" because we want to read in binary mode
    bird_sim1 = dill.load(pickle_in)
```

```python
def plot_va_t(va_matrix, shape, color, labels, index):
    # Plot order parameter as a function of time for two systems: one reachs phase transition and one does not
    plt.title("Evolution of Order Parameter")
    plt.xlabel("t")
    plt.ylabel("$v_a$")
    plt.plot(np.arange(0, len(va_matrix[index]), 1), va_matrix[index], shape, markerfacecolor='none', markeredgecolor=color, label=labels)
    plt.legend()
      
plot_va_t(bird_sim1.va_matrix, 'o','#b68cb8','N = 40 , η = 0', 1) 
plot_va_t(bird_sim1.va_matrix, 's','#314cb6','N = 40, η = 3',30)
plot_va_t(bird_sim1.va_matrix, '^','#b33951','N = 40, η = 5', -1)
print(bird_sim1.eta_matrix[8]) 
# i just tried some values to get three different plots and 8 seemd good but if you dnt want it we can delete it
plt.show()
```
